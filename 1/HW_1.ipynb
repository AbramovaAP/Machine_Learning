# Урок 1. Функция потерь. Загрузка датасетов    
Вам дан набор данных с фактическими значениями целевой переменной и предсказанными значениями:


**Фактические значения:**    
actual_values = np.array([100, 150, 200, 250])

**Предсказанные значения:**    
predicted_values = np.array([90, 140, 180, 270])

Задача студентов состоит в расчете среднеквадратичной ошибки (MSE) и средней абсолютной ошибки (MAE) для данного набора данных без использования модели. Студенты должны использовать правильные значения фактической переменной и предсказанные значения для расчета функций потерь. Они также должны проанализировать результаты и объяснить, какие значения функции потерь указывают на лучшую точность предсказания.

```python
import numpy as np

# Фактические значения
actual_values = np.array([100, 150, 200, 250])

# Предсказанные значения
predicted_values = np.array([90, 140, 180, 270])

# Расчет MSE
mse = np.mean((actual_values - predicted_values) ** 2)

# Расчет MAE
mae = np.mean(np.abs(actual_values - predicted_values))

# Вывод результатов
print(f"Среднеквадратичная ошибка (MSE): {mse}") 
print(f"Средняя абсолютная ошибка (MAE): {mae}")
```

```python
# Вариант № 2 - Без numpy
# Фактические и предсказанные значения
actual_values = [100, 150, 200, 250] 
predicted_values = [90, 140, 180, 270]

# Вычисление MSE
mse = sum((a - p) ** 2 for a, p in zip(actual_values, predicted_values)) / len(actual_values)

# Вычисление MAE
mae = sum(abs(a - p) for a, p in zip(actual_values, predicted_values)) / len(actual_values)

# Вывод результатов
print(f"Среднеквадратичная ошибка (MSE): {mse}") 
print(f"Средняя абсолютная ошибка (MAE): {mae}")
```

## Анализ результатов    
После выполнения кода получаем значения MSE и MAE.    

**MSE:** Чем меньше значение MSE, тем лучше качество модели, так как это значение указывает на среднее квадратичное отклонение предсказанных значений от фактических. Если MSE равно нулю, это означает, что предсказанные значения полностью совпадают с фактическими.

**MAE:** Этот показатель также показывает, насколько близки предсказанные значения к фактическим. MAE, как и MSE, должен быть минимальным для достижения хорошего качества предсказания. Если MAE равно нулю, это также означает идеальное совпадение предсказаний и фактических значений.

## Заключение
Оба показателя — MSE и MAE — используются для оценки точности предсказаний. Однако MSE более чувствителен к выбросам, так как квадраты ошибок увеличивают влияние больших отклонений. MAE, в свою очередь, более устойчив к выбросам, так как учитывает абсолютные значения ошибок. В зависимости от задачи можно выбирать тот или иной показатель для оценки качества модели.
